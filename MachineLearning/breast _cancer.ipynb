## Training the Model

```python
df1['target']=df.target
```

## Train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=101)
```

## standard Scalar

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train1=sc.fit_transform(x_train)
x_train_scaled=pd.DataFrame(x_train1)
x_train_scaled.columns=df.feature_names
x_test1=sc.transform(x_test)
x_test_scaled=pd.DataFrame(x_test1)
x_test_scaled.columns=df.feature_names
print(x_train_scaled)
type(x_test_scaled)
```

## RFE(Recursive Feature Elimination)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
```

```python
lgr=LogisticRegression()
rfe=RFE(lgr,15)
rfe=rfe.fit(x_train_scaled,y_train)
```

## kmeans

```python
from sklearn.cluster import KMeans
cluster_id=[]
k=list(range(1,7))
for i in range(1,7):
    km=KMeans(i).fit(x_train)
    y_km_pred=km.fit_predict(x_train)
    cluster_id.append(km.inertia_)
print(cluster_id)
print(km.labels_)
```

```python
plt.plot(k,cluster_id)
```

```python
Km=KMeans(n_clusters=2,max_iter=50,random_state=42).fit(x_train)
Km
```

```python
Km.cluster_centers_
```

```python
y_means=Km.fit_predict(x_train)
y_means
```

```python
x = x_train.iloc[:, [3, 4]].values
x
```

```python
km = KMeans(n_clusters = 2, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)

plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'green', label = 'cluster1')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'yellow', label = 'cluster2')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'black' , label = 'centeroid')

plt.legend()
plt.show()
```
