## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_new1['Churn']=le.fit_transform(df_new1['Churn'])
```

```python
df_new1
```

## converting object column to numeric

```python
df_new1['TotalCharges']=pd.to_numeric(df_new1['TotalCharges'],errors='correc')
df_new1.info()
df_new1.isnull().sum()
df_new1['TotalCharges'].fillna(value=df_new1['TotalCharges'].mean(),inplace=True)
```

```python
df_new1['PhoneService']=le.fit_transform(df_new1['PhoneService'])
df_new1['Contract']=le.fit_transform(df_new1['Contract'])
df_new1['PaperlessBilling']=le.fit_transform(df_new1['PaperlessBilling'])
df_new1['PaymentMethod']=le.fit_transform(df_new1['PaymentMethod'])
df_new1['gender']=le.fit_transform(df_new1['gender'])
df_new1['Partner']=le.fit_transform(df_new1['Partner'])
df_new1['Dependents']=le.fit_transform(df_new1['Dependents'])
df_new1['MultipleLines']=le.fit_transform(df_new1['MultipleLines'])
```

```python
df_new1
```

```python
final=df_new1.drop('customerID',1)
```

```python
final
```

```python
d1=pd.get_dummies(final,drop_first=True)
```

```python
d1
```

```python
d1.info()
```

```python
d1.skew()
```

```python
y=d1['Churn']
x=d1.drop('Churn',1)
```

## Training the model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=101)
```

## Logistic Regression

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
input=[('scale',StandardScaler()),('mode',LogisticRegression())]
print(input)
pipe=Pipeline(input)
res=pipe.fit(x_train,y_train)
yhat=res.predict(x_train)
print(yhat)
print(accuracy_score(y_train,yhat))
print(classification_report(y_train,yhat))
```

```python

yhat=res.predict(x_test)
print(yhat)
print(accuracy_score(y_test,yhat))
print(classification_report(y_test,yhat))
```

```python
sns.regplot(y_test,yhat)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
input=[('scale',StandardScaler()),('mode',KNeighborsClassifier(n_neighbors=3))]
print(input)
pipe=Pipeline(input)
res=pipe.fit(x_train,y_train)

yhat=res.predict(x_train)
print(yhat)
print(accuracy_score(y_train,yhat))
print(classification_report(y_train,yhat))
```

```python
yhat=res.predict(x_test)
print(yhat)
print(accuracy_score(y_test,yhat))
print(classification_report(y_test,yhat))
```

## SVC

```python
from sklearn.svm import SVC
```

```python
from sklearn.preprocessing import MinMaxScaler
svc=SVC(C=10,kernel='linear',gamma=0.001,degree=1)
input=[['scale',MinMaxScaler()],['mode',svc]]
print(input)
pipe=Pipeline(input)
res=pipe.fit(x_train,y_train)

yhat=res.predict(x_test)
print(yhat)
print(accuracy_score(y_test,yhat))
print(classification_report(y_test,yhat))
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB
from sklearn.preprocessing import StandardScaler
gb=GaussianNB()
br=BernoulliNB()
input=[['scale',StandardScaler()],['mode',br]]
print(input)
pipe=Pipeline(input)
res=pipe.fit(x_train,y_train)

yhat=res.predict(x_test)
print(yhat)
print(accuracy_score(y_test,yhat))
print(classification_report(y_test,yhat))
```
