## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x_train=pd.read_csv('gisette_train.data',delimiter=' ')
x_train
```

```python
x_train=x_train.drop('Unnamed: 5000',1)
```

```python
y_train=pd.read_csv('gisette_train.labels')
```

```python
y_train
```

```python
x_test=pd.read_csv('gisette_valid.data',delimiter=' ')
x_test
```

```python
y_test=pd.read_csv('gisette_valid.labels')
```

```python
y_test
```

```python
x_test.columns
```

```python
x_test=x_test.drop('Unnamed: 5000',1)
```

```python
y_test
```

## Training the model with PCA

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
pca=PCA(n_components=20)
input=[['scale',StandardScaler()],['mode',pca]]
pipe=Pipeline(input)
res1=pipe.fit_transform(x_train)
res2=pipe.transform(x_test)
print(res1)
```

```python
XprincipalDf = pd.DataFrame(res1)
XprincipalDf_t = pd.DataFrame(res2)
```

## Logistic Regression

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
input=[('mode',LogisticRegression())]
print(input)
pipe=Pipeline(input)
res=pipe.fit(XprincipalDf,y_train)

yhat=res.predict(XprincipalDf)
print(yhat)
print(accuracy_score(y_train,yhat))
print(classification_report(y_train,yhat))
```

```python

yhat=res.predict(XprincipalDf_t)
print(yhat)
print(accuracy_score(y_test,yhat))
print(classification_report(y_test,yhat))
```

## KNN

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
input=[('mode',KNeighborsClassifier(n_neighbors=7))]
print(input)
pipe=Pipeline(input)
res=pipe.fit(XprincipalDf,y_train)

yhat=res.predict(XprincipalDf)
print(yhat)
print(accuracy_score(y_train,yhat))
print(classification_report(y_train,yhat))
```

```python

yhat=res.predict(XprincipalDf_t)
print(yhat)
print(accuracy_score(y_test,yhat))
print(classification_report(y_test,yhat))
```
