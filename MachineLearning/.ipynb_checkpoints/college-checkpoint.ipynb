## Importing the libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("College_Data")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
```

```python
df=df.drop('Unnamed: 0',1)
```

## Get_dummies

```python
df_new=pd.get_dummies(df)
df_new
```

## Kmeans 

```python
from sklearn.cluster import KMeans
cluster_id=[]
k=list(range(1,7))
for i in range(1,7):
    km=KMeans(i).fit(df_new)
    y_km_pred=km.fit_predict(df_new)
    cluster_id.append(km.inertia_)
print(cluster_id)
print(km.labels_)
```

```python
plt.plot(k,cluster_id)
```

## Kmeans (pipeline method)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
input=[('scale',StandardScaler()),('mode',KMeans(n_clusters=2,init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0))]
print(input)
pipe=Pipeline(input)
res=pipe.fit(df_new)
y_means=res.predict(df_new)
print(y_means)
```

```python
x = df_new.iloc[:, [7,8]].values
x
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'green', label = 'cluster1')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'yellow', label = 'cluster2')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s = 50, c = 'cyan' , label = 'centeroid')




plt.legend()
plt.show()
```

## Zscore 

```python
from scipy import stats
z=np.abs(stats.zscore(df_new))
df_new1=df_new[(z<3).all(axis=1)]
df_new1
```

```python
df_new1.skew()
```

```python
from sklearn.cluster import KMeans
cluster_id=[]
k=list(range(1,7))
for i in range(1,7):
    km=KMeans(i).fit(df_new1)
    y_km_pred=km.fit_predict(df_new1)
    cluster_id.append(km.inertia_)
print(cluster_id)
print(km.labels_)
```

```python
plt.plot(k,cluster_id)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
input=[('scale',StandardScaler()),('mode',KMeans(n_clusters=3,init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0))]
print(input)
pipe=Pipeline(input)
res=pipe.fit(df_new1)
y_means=res.predict(df_new1)
print(y_means)
```

```python
x = df_new1.iloc[:, [7,8]].values
x


plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'green', label = 'cluster1')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'yellow', label = 'cluster2')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'red', label = 'cluster2')


plt.legend()
plt.show()
```
