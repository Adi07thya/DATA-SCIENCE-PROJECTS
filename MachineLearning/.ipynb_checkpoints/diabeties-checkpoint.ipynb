## importing the libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import numpy as np
warnings.filterwarnings('ignore')
from sklearn.datasets import load_diabetes
```

```python
dia=load_diabetes()
```

```python
dia
```

```python
dia.feature_names
```

```python
df=pd.DataFrame(dia.data)
df.columns=dia.feature_names
df['target']=dia.target
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.skew()
```

## Traing the model

```python
x=df.drop('target',1)
y=df[['target']]
```

## train_test_split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.4,random_state=101)

```

```python
parameters={'C':[0.001,0.01,.1,1,10,100,1000,10000],'kernel':['linear','rbf'],'gamma':[0.001,0.01,.1,1,10],'degree':[1,2,3,4]}
```

## SVR using Gridsearch cv

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV 
grid = GridSearchCV(SVR(),parameters) 
grid.fit(X_train, Y_train) 
print(grid.best_estimator_)
print(grid.best_params_)
```

## SVC

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score,classification_report
svc=SVR(C=1000,kernel='rbf',gamma=0.1,degree=1)
input=[['scale',MinMaxScaler()],['mode',svc]]
print(input)
pipe=Pipeline(input)
res=pipe.fit(X_train,Y_train)
yhat=res.predict(X_train)

data=pd.DataFrame(yhat)
```

```python
from sklearn.metrics import accuracy_score,classification_report,r2_score
```

```python
r2_score(Y_train,yhat)
```

```python
sns.regplot(Y_train,yhat)
```
