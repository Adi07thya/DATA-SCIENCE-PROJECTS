## converting object column to integer column

```python
df[' 125']=df.apply(pd.to_numeric,errors='correc')
df[' 125.1']=df.apply(pd.to_numeric,errors='correc')
df['   1.0']=df.apply(pd.to_numeric,errors='correc')
df[ '1']=df.apply(pd.to_numeric,errors='correc')
```

##  Replacing ? to nan

```python
df=df.replace('   ?',np.nan)
df=df.replace('     ?',np.nan)
```

## Training the model

```python
df2=df1.drop('ad.',1)
```

## Principal Component Analysis(PCA)-Pipeline method

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
pca=PCA(n_components=20)
input=[['scale',StandardScaler()],['mode',pca]]
pipe=Pipeline(input)
res1=pipe.fit_transform(df2)
res1
print(res1)


```

```python
principalDf = pd.DataFrame(res1)
```

```python
principalDf
```

```python
principalDf
finalDf = pd.concat([principalDf, df[['ad.']]], axis = 1)
```

```python
finalDf
```

```python
x1=finalDf.drop('ad.',1)
y2=finalDf['ad.']
```

## Train_test_split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x1,y2,test_size=0.3,random_state=101)
```

## Logistic Regression (Pipeline method)

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
input=[('mode',LogisticRegression())]
print(input)
pipe=Pipeline(input)
res=pipe.fit(X_train,Y_train)

yhat=res.predict(X_train)
print(yhat)
print(accuracy_score(Y_train,yhat))
print(classification_report(Y_train,yhat))
```

```python
pipe=Pipeline(input)
res=pipe.fit(X_train,Y_train)
print(res)
yhat=res.predict(X_test)
print(yhat)
print(accuracy_score(Y_test,yhat))
print(classification_report(Y_test,yhat))
```

## KNeighbors classifiers(pipeline method)

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
input=[('mode',KNeighborsClassifier(n_neighbors=5))]
print(input)
pipe=Pipeline(input)
res=pipe.fit(X_train,Y_train)

yhat=res.predict(X_train)
print(yhat)
print(accuracy_score(Y_train,yhat))
print(classification_report(Y_train,yhat))
```

```python
yhat=res.predict(X_test)
print(yhat)
print(accuracy_score(Y_test,yhat))
print(classification_report(Y_test,yhat))
```
